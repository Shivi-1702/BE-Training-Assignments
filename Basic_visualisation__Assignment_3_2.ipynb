```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("Iris.csv")
```

```python
Data
```

```python
Data.head(50)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Data.index, Data['SepalLengthCm'], marker='o', linestyle='-', color='black')
```

```python
plt.plot(Data['SepalLengthCm'],Data['PetalLengthCm'])
```

```python
x = Data['SepalWidthCm']
y = Data['PetalWidthCm']

plt.plot(x, y, marker=',', linestyle='-', color='y')

plt.xlabel('SepalWidthCm')
plt.ylabel('PetalWidthCm')
plt.title('Species')
```

```python
plt.scatter(Data ['SepalLengthCm'], Data ['SepalWidthCm'])

plt.xlabel("Sepal Width")
plt.ylabel('Sepal Length')
```

```python
plt.scatter(x,y,marker = "*",color = "green")
```

```python
  plt.scatter(Data['SepalLengthCm'],Data['SepalWidthCm'], alpha=0.7, edgecolor='pink')
  plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.bar(Data['Species'], Data['PetalLengthCm'])
plt.grid()
```

```python
plt.barh(Data ['Species'], Data ['PetalLengthCm'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Species')
plt.title('Petal Length by Species')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
labels = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

sizes = [Data ['SepalLengthCm'].mean(), Data ['SepalWidthCm'].mean(), Data ['PetalLengthCm'].mean(), Data['PetalWidthCm'].mean()]
colors = ['red', 'green', 'yellow', 'blue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.f%%', startangle=140)

plt.title('pie chart')

plt.show()
```
