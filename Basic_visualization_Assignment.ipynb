```python
#import
import matplotlib.pyplot as plt
```

```python
# Average Temperature Over Years

Year = [2010, 2011, 2012, 2013, 2014]
Temperature = [22, 23, 21, 24, 25]

# Create a plot
plt.plot(Year, Temperature, marker='o', linestyle='-', color='b', label='Prime Numbers')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()

# Show the plot
plt.show()
```

```python
# Height vs Weight

Height = [150, 160, 165, 170, 175, 180]
Weight = [50, 55, 60, 65, 70, 75]

plt.scatter(Height, Weight, marker = "v",color = "red")
```

```python
# Number of Students per Subject

Subjects = ['Math', 'Science', 'English', 'History']
Students = [30, 25, 35, 20]

import matplotlib.pyplot as plt

# Create a bar chart
plt.bar(Subjects, Students, color='skyblue')

# Add labels and title
plt.xlabel('Names')
plt.ylabel('Paisa')
plt.title('Bar Chart Example')

# Show the plot

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Daily Activity Distribution

Activities = ['Sleep', 'Study', 'Exercise', 'Entertainment']
Hours = [8, 6, 2, 8]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Create a pie chart
plt.pie(Hours, labels=Activities, colors=colors, autopct='%1.f%%', startangle=140)

# Add title
plt.title('Daily Activity Distribution')

# Show the plot
plt.show()
```
