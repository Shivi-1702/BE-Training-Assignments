```python
import pandas as pd
```

```python
df = pd.read_csv('car_price_dataset.csv')
```

```python
df
```

```python
# What are all the unique car brands (make) in the dataset?

df['Brand']
```

```python
# What are the different types of fuel available?

df['Fuel_Type'].unique()
```

```python
# Find all cars that use Diesel fuel.

df[df['Fuel_Type']=="Diesel"]
```

```python
# What are the different types of transmission in the dataset?

df['Transmission'].unique()
```

```python
# Find all cars with automatic transmission.

df[df['Transmission']=="Automatic"]
```

```python
# Find all cars manufactured in the year 2015

df[df['Year']=="2015"]
```

```python
# Find all cars priced less than $10,000.

df[df['Price']<10000]
```

```python
# Find all cars that have driven less than 50,000 kilometers (or miles).

df[df['Mileage']<50000]
```

```python
# Find all cars that are of body type 'SUV'

df[df['Fuel_Type']=="SUV"]
```
